**`Evaluasi Metrik`**

Miss-prediction terjadi ketika model gagal memprediksi jumlah penyewaan sepeda dengan akurat. Hal ini bisa bermanifestasi sebagai over-prediction, di mana model memprediksi permintaan yang lebih tinggi dari yang sebenarnya, atau under-prediction, di mana model memprediksi permintaan yang lebih rendah dari yang sebenarnya.

- **Over-prediction:** Dapat menyebabkan surplus sepeda di beberapa lokasi, mengakibatkan sumber daya tidak digunakan secara efisien, meningkatkan biaya operasional dan pemeliharaan. Misalnya, jika 300 sepeda disiapkan untuk disewa pada satu hari, tetapi permintaan hanya untuk 150 sepeda selama hari tersebut, maka jika biaya operasional (misalnya, pemeliharaan) adalah 0.50 USD per sepeda per hari, kerugian biaya operasional adalah (300 - 150) * 0.50 USD = 75 USD. [1](https://www.businessanalystlearnings.com/business-matters/2022/2/3/how-much-does-it-cost-to-implement-a-bike-share-system-in-2022)
- **Under-prediction:** Berisiko menyebabkan kekurangan sepeda, menurunkan kepuasan pelanggan dan kehilangan peluang pendapatan karena pengguna tidak dapat menemukan sepeda saat dibutuhkan. Misalnya, dalam skenario penyewaan sepeda selama 30 menit untuk pengguna biasa pada sepeda klasik, biayanya adalah 1 USD + (0.05 USD * 30) = 2.5 USD. Sekarang, jika situasi di mana tersedia 80 sepeda untuk disewa, tetapi permintaan sebenarnya adalah 120 sepeda pada jam tertentu. Dalam kasus ini, potensi pendapatan yang hilang dalam satu jam adalah (120 - 80) * 2.5 USD = 100 USD. [2](https://capitalbikeshare.com/pricing)

Metrik Evaluasi yang digunakan

| Metric | Penjelasan |
|--------|-------------|
| **MAE (Mean Absolute Error)** | MAE mengukur rata-rata dari nilai absolut kesalahan antara prediksi dan nilai sebenarnya. Ini memberikan pemahaman langsung tentang kesalahan rata-rata yang diharapkan dalam prediksi jumlah sepeda. MAE sangat berguna untuk menilai dampak rata-rata dari miss-predictions dan membantu dalam optimasi persediaan karena tidak memberikan hukuman berlebihan untuk outlier. |
| **MAPE (Mean Absolute Percentage Error)** | MAPE sangat berguna karena mengukur kesalahan sebagai persentase dari nilai sebenarnya, yang sangat relevan dalam kasus di mana fluktuasi permintaan memiliki skala yang luas sepanjang tahun (berdasarkan musim atau kondisi cuaca). Ini memberikan perspektif tentang seberapa besar kesalahan prediksi relatif terhadap jumlah sepeda yang sebenarnya disewakan. |
| **R² (Coefficient of Determination)** | R² bisa menunjukkan seberapa baik model dapat menjelaskan variabilitas permintaan sepeda. Nilai yang tinggi menunjukkan bahwa model dengan akurat menjelaskan perubahan dalam jumlah penyewaan sepeda dan berpotensi dapat dipercaya dalam setting operasional untuk prediksi permintaan. |

Dengan menggunakan kombinasi ketiga metrik ini, kita dapat memperoleh gambaran menyeluruh tentang kinerja model. Metrik ini membantu menilai seberapa baik model memenuhi kebutuhan operasional penyewaan sepeda, mengoptimalkan persediaan, dan meminimalkan biaya serta memaksimalkan kepuasan pelanggan. Evaluasi ini selaras dengan goals untuk mengembangkan model yang dapat secara akurat memprediksi permintaan penyewaan sepeda dan mendukung pengambilan keputusan strategis dalam manajemen sumber daya.

**`Keterbatasan Proyek`**

Setiap proyek analitik menghadapi batasan tertentu yang dapat mempengaruhi ruang lingkup dan hasil akhir dari inisiatif tersebut. Untuk model prediksi penyewaan sepeda, batasan-batasan ini mencakup:

1. Ketersediaan Data:
- Proyek ini terbatas pada data yang sudah tersedia yang mencakup tanggal, waktu, cuaca, dan jumlah penyewaan sepeda. Keterbatasan data seperti ketidaktersediaan informasi spesifik lokasi atau data perilaku pengguna dapat membatasi akurasi dan aplikabilitas model dalam memprediksi permintaan di lokasi spesifik atau untuk segmentasi pengguna tertentu.
2. Kualitas Data:
- Data yang digunakan diharapkan bersih dan telah melalui proses pembersihan awal; namun, kemungkinan masih adanya noise atau outliers yang tidak terdeteksi. Hal ini dapat menyebabkan model prediktif mengalami bias atau overfitting, yang mengurangi kemampuan generalisasi model pada data baru atau dalam kondisi operasional yang berbeda.
3. Kompleksitas Model:
- Model yang dipilih harus seimbang antara akurasi prediksi dan kompleksitas komputasi. Model yang terlalu kompleks mungkin memerlukan sumber daya komputasi yang signifikan, yang bisa tidak praktis untuk implementasi operasional real-time.
4. Variabilitas Kondisi Eksternal:
- Model akan menggunakan data historis cuaca, namun perubahan iklim dan variabilitas cuaca tahunan mungkin tidak sepenuhnya terprediksi. Ini bisa mengakibatkan model kurang akurat dalam merespons kondisi cuaca yang tidak terduga atau perubahan pola cuaca jangka panjang.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
# Load Data
main = pd.read_csv('../data/raw/data_bike_sharing.csv', parse_dates=['dteday'])
add = pd.read_csv('../data/extend/bike_sharing_hourly.csv', parse_dates=['dteday'])

merge = pd.merge(main, add[['dteday', 'hr', 'windspeed']], on=['dteday', 'hr'], how='left')
merge = merge[['dteday', 'hr', 'holiday', 'season', 'weathersit', 'hum', 'atemp', 'temp', 'windspeed', 'casual', 'registered', 'cnt']]
data = merge.copy()

print('Dataset utama:',main.shape)
display(main.head())

print('Dataset tambahan:',add.shape)
display(add.head())

print('Dataset gabungan:',data.shape)
display(data.head())
```

```python
listItem = []

# Loop through each column to gather descriptive statistics
for col in data.columns:
    data_type = data[col].dtype
    missing_percentage = data[col].isna().mean() * 100
    if data_type in [np.int64, np.float64]:
        negative_percentage = (data[col] < 0).mean() * 100
    else:
        negative_percentage = 0
    duplicate_count = data.duplicated(subset=col).sum()
    unique_count = data[col].nunique()
    unique_sample = data[col].unique()

    listItem.append([
        col, 
        data_type, 
        missing_percentage, 
        negative_percentage, 
        duplicate_count, 
        unique_count, 
        unique_sample
    ])

# Create DataFrame from the collected descriptions
tabelDesc = pd.DataFrame(listItem, columns=[
    'Column Name', 
    'Data Type', 
    'Missing Value(%)', 
    'Negative Value (%)', 
    'Duplicate', 
    'Number of Unique', 
    'Unique Sample'
])

# Display the descriptive table
tabelDesc
```

* Tidak ada missing value, nilai negatif, dan duplikat data (0.0%) di seluruh kolom. Ini menunjukkan bahwa data lengkap tanpa missing values dan duplikasi, serta data dalam setiap kolom berada dalam rentang yang diharapkan tanpa anomali negatif.

**Tes Normalitas Shapiro**

```python
from scipy.stats import shapiro

dist = []

for i in range(1, len(data.columns)):
    column_names = data.columns[i]
    stat, p_value = shapiro(data[column_names])
    
    if p_value > 0.05:
        distribution = "True"
    else:
        distribution = "False"
    
    dist.append([column_names, stat, p_value, distribution])

dist_df = pd.DataFrame(dist, columns=['Columns', 'Statistic', 'P-value', 'Normal Distribution'])
dist_df
```

Semua fitur, baik numerik maupun kategorikal yang telah dienkode sebagai angka, **tidak berdistribusi normal karena semua nilai p lebih kecil dari 0.05**.

Untuk langkah selanjutnya, kita menghitung skor IQR untuk menentukan secara numerik outlier pada kolom yang relevan seperti `hum`, `atemp`, `windspeed`, `temp`, `casual`, `registered`, dan `cnt`.

```python
# Fungsi untuk menghitung dan mengembalikan outlier menggunakan metode IQR
def detect_outliers_iqr(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter outliers
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

# Menentukan outlier untuk setiap kolom numerik yang dipilih
outlier_columns = ['hum', 'atemp', 'temp', 'casual', 'registered', 'windspeed', 'cnt']
outliers_dict = {col: detect_outliers_iqr(data_dummy, col) for col in outlier_columns}

# Jumlah outlier di setiap kolom
outliers_count = {col: len(outliers_dict[col]) for col in outlier_columns}
outliers_count
```

**3. Feature Engineering**

Feature engineering adalah proses kreatif dalam machine learning yang melibatkan pengubahan data mentah menjadi fitur yang dapat digunakan untuk membuat model prediktif yang efektif. 

Mengkategorikan data `hr` (dalam format 24-jam) menjadi beberapa sesi waktu dalam sehari. Kategorisasi ini dapat membantu dalam analisis data dengan membagi waktu menjadi segmen-segmen yang lebih mudah dikelola dan dianalisis. Dengan demikian, pola-pola atau tren tertentu bisa lebih mudah diidentifikasi berdasarkan waktu dalam sehari.

```python
def session(x):
    
    if x>4 and x<=8:
        return 0
    elif x>8 and x<=12:
        return 1
    elif x>12 and x<=16:
        return 2
    elif x>16 and x<=20:
        return 3
    elif x>20 and x<=24:
        return 4
    elif x<=4:
        return 5

#apply funtion to make new category column
data_dummy['session'] = data_dummy['hr'].apply(session)
data_dummy.head()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from scipy.stats.mstats import winsorize
from typing import Literal

class OutlierHandling(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self 
    
    def transform(self, X, y=None):
        # Handling outliers in 'windspeed' column
        if 'windspeed' in X.columns:
            X['windspeed'] = winsorize(X['windspeed'], limits=(0.01, 0.01))
        
        # Handling outliers in 'hum' column
        if 'hum' in X.columns:
            medianHum = X.groupby(['weathersit','season'])['hum'].median()
            X['hum'] = X.groupby(['season', 'weathersit'])['hum'].transform(lambda x: x.replace(0, x.median()))
        
        # Replace 4 with 3 in 'weathersit' column
        if 'weathersit' in X.columns:
            X['weathersit'] = X['weathersit'].replace(4, 3)
        
        return X
    
    def set_output(self, transform: Literal['default', 'pandas']):
        return super().set_output(transform=transform)

class FeatureCreation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    def session(self, x):
        if x > 4 and x <= 8:
            return 0
        elif x > 8 and x <= 12:
            return 1
        elif x > 12 and x <= 16:
            return 2
        elif x > 16 and x <= 20:
            return 3
        elif x > 20 and x <= 24:
            return 4
        elif x <= 4:
            return 5
    
    def transform(self, X, y=None):
        X['session'] = X['hr'].apply(self.session)
        
        return X
```

```python
from pycaret.internal.preprocess.transformers import TransformerWrapper
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler, OneHotEncoder, PowerTransformer, StandardScaler
from category_encoders import BinaryEncoder
from pycaret.internal.preprocess.transformers import ExtractDateTimeFeatures
from mlxtend.feature_selection import ColumnSelector

outliers_handling = TransformerWrapper(
                    include=['windspeed', 'hum', 'weathersit', 'season'],
                    transformer=OutlierHandling())

feature_creation = TransformerWrapper(
        include=['hr'], 
        transformer=FeatureCreation())

extract_date = TransformerWrapper(
                include=['dteday'],
                transformer=ExtractDateTimeFeatures(
                    features=['day', 'month', 'weekday']))

selection = ColumnSelector(cols=['dteday_day', 'dteday_month', 'dteday_weekday', 'hr', 'holiday', 'season', 'weathersit', 'hum', 'temp', 'windspeed', 'session'])

encoder = TransformerWrapper(
        include=['hr', 'dteday_day', 'dteday_month', 'dteday_weekday', 'season', 'weathersit', 'session'],
        transformer=OneHotEncoder(drop='first', handle_unknown='ignore'))

preprocessor = [
    ('extract date', extract_date),
    ('feature_creation', feature_creation),
    ('selection', selection),
    ('outlier handling', outliers_handling)
]
```

**Perbandingan K-Fold dari Cross-Validation**

```python
import time

cross_val = pd.DataFrame()

for k in range(1, 11):
    start_time = time.time()
    if k == 1:
        model = create_model(tuned_models['catboost'], verbose=False, cross_validation=False)
        results = pd.DataFrame(pull().loc['Test']).T
    else:
        model = create_model(tuned_models['catboost'], verbose=False, fold=k)
        results = pd.DataFrame(pull().loc['Mean']).T

    results.insert(0, 'K-Fold', k)
    elapsed_time = time.time() - start_time
    results['Time (Sec)'] = elapsed_time
    cross_val = cross_val.append(results, ignore_index=True)

cross_val
```

```python
sns.set_style('darkgrid')

fig, axs = plt.subplots(2, 3, figsize =(20,8))
axs = axs.flatten()

metric_eval = ['MAE', 'RMSE', 'RMSLE', 'MAPE', 'R2', 'Time (Sec)']
for i in range(len(metric_eval)):
    sns.lineplot(data=cross_val, x='K-Fold', y=metric_eval[i], ax=axs[i], marker='o')

fig.tight_layout()
plt.suptitle('K-Fold Cross-Validation Results', y=1.02, fontsize=16)
plt.show()
```

Learning Curve dengan MAPE membantu memahami bagaimana model belajar dari data dan bagaimana performanya pada data yang tidak terlihat. - Skor pada data pelatihan (garis biru) secara konsisten meningkat seiring dengan bertambahnya jumlah instance pelatihan.

**Cara Kerja Model CatBoost**

CatBoost, singkatan dari "Categorical Boosting," adalah algoritma pembelajaran mesin berbasis boosting yang dikembangkan oleh Yandex. Algoritma ini dirancang untuk bekerja dengan baik pada data kategorikal tanpa memerlukan banyak praproses. CatBoost menggunakan pendekatan gradient boosting pada pohon keputusan (decision trees) untuk membangun model prediktif yang kuat.

**Tahapan Utama dalam Cara Kerja CatBoost:**

1. **Boosting**:
   - CatBoost bekerja dengan pendekatan boosting, di mana model dibangun secara iteratif. Pada setiap iterasi, model baru dibuat untuk mengoreksi kesalahan model sebelumnya. Model ini menggabungkan banyak pohon keputusan yang lemah menjadi satu model yang kuat.
   
2. **Handling Categorical Features**:
   - Salah satu fitur utama dari CatBoost adalah kemampuannya untuk menangani data kategorikal dengan efisien. Algoritma ini mengubah fitur kategorikal menjadi nilai numerik menggunakan metode yang disebut "ordered boosting," yang mengurangi bias informasi masa depan dalam pengolahan data pelatihan.

3. **Ordered Boosting**:
   - Ordered boosting adalah teknik yang digunakan oleh CatBoost untuk memastikan bahwa saat model dilatih, data pelatihan diurutkan sedemikian rupa sehingga tidak terjadi kebocoran informasi antara pengamatan yang digunakan untuk membangun model dan pengamatan yang digunakan untuk mengukur performa model.
   
4. **Gradient-Based Learning**:
   - Seperti algoritma gradient boosting lainnya, CatBoost menggunakan gradien dari fungsi kerugian untuk membimbing pembangunan pohon keputusan berikutnya. Pohon baru dibangun untuk meminimalkan kesalahan residu dari pohon sebelumnya.

5. **Regularization and Overfitting Control**:
   - CatBoost memiliki berbagai mekanisme untuk menghindari overfitting, termasuk penggunaan regularisasi L2, pemangkasan pohon (tree pruning), dan penyesuaian bobot secara hati-hati selama proses boosting.

6. **Speed and Efficiency**:
   - CatBoost dirancang untuk efisiensi waktu dan penggunaan sumber daya. Algoritma ini mengimplementasikan berbagai optimisasi untuk mempercepat waktu pelatihan dan prediksi, seperti pemrosesan paralel dan komputasi berbasis GPU.

7. **Loss Function**:
   - CatBoost mendukung berbagai fungsi kerugian (loss function) yang dapat digunakan untuk berbagai jenis masalah, termasuk regresi, klasifikasi biner, dan klasifikasi multi-kelas.

8. **Model Interpretation**:
   - Selain performa yang kuat, CatBoost juga menyediakan alat untuk interpretasi model, seperti pentingnya fitur (feature importance) dan plot SHAP (Shapley Additive Explanations), yang membantu pengguna memahami bagaimana model membuat prediksi.

Dengan menggabungkan kekuatan gradient boosting, kemampuan penanganan data kategorikal yang efisien, dan teknik regulasi untuk menghindari overfitting, CatBoost memberikan solusi yang kuat dan andal untuk berbagai jenis masalah pembelajaran mesin, dari prediksi nilai kontinu hingga klasifikasi kategori.

```python
def add_error(data, error_percentage):
    return data * (1 + error_percentage / 100)

error_percentages = [10, 20, 30]

# Menambahkan error untuk skenario overpredict dan underpredict
scenarios = {}
for error in error_percentages:
    scenarios[f'over_noML{error}%'] = add_error(prediction['cnt'], error)
    scenarios[f'under_noML{error}%'] = add_error(prediction['cnt'], -error)

df_scenarios = pd.DataFrame(scenarios)
df_results = pd.concat([prediction, df_scenarios], axis=1)
df = df_results.round().astype(int)
display(df.head())
```

```python
bins = [0, 50, 100, 200, 300, 400, 500, 600, float('inf')]
labels = ['0-50', '50-100', '101-200', '201-300', '301-400', '401-500', '501-600', '>600']
df['bin'] = pd.cut(df['cnt'], bins=bins, labels=labels)
```

**Asumsi**
- Biaya Maintenance Sepeda: 0.55 USD per sepeda per hari.
- Pendapatan per Pengguna (Single Ride, [Capital Bike Share](https://capitalbikeshare.com/pricing)):
    - Biaya Unlock: 1 USD.
    - Biaya per menit: 0.05 USD.
- Rata-rata Durasi Perjalanan: 30 menit per hari.
- Total Pendapatan per Pengguna: 1 USD + (0.05 USD x 30 menit) = 2.5 USD per pengguna.

```python
def calculate_costs_per_bin(df, prediction_col, actual_col='cnt'):
    maintenance_cost_per_bike = 0.55
    revenue_per_user = 2.5
    
    cost_data = []

    for label in labels:
        bin_df = df[df['bin'] == label]
        overpredict_cost = 0
        underpredict_cost = 0
        
        for idx, row in bin_df.iterrows():
            predicted = row[prediction_col]
            actual = row[actual_col]
            
            if predicted > actual:
                overpredict_bikes = predicted - actual
                overpredict_cost += overpredict_bikes * maintenance_cost_per_bike
            elif predicted < actual:
                underpredict_users = actual - predicted
                underpredict_cost += underpredict_users * revenue_per_user
        
        cost_data.append((label, overpredict_cost, underpredict_cost))
    
    return pd.DataFrame(cost_data, columns=['bin', 'overpredict_cost', 'underpredict_cost'])

scenarios = ['prediction_label'] + [f'over_noML{error}%' for error in error_percentages] + [f'under_noML{error}%' for error in error_percentages]
bin_results = {}

for scenario in scenarios:
    bin_results[scenario] = calculate_costs_per_bin(df, scenario)

# Menggabungkan hasil ke dalam satu DataFrame
df_bin_summary = pd.concat(bin_results, axis=1)
df_bin_summary.columns = [' '.join(col).strip() for col in df_bin_summary.columns.values]
df_bin_summary = df_bin_summary[['prediction_label bin', 'prediction_label overpredict_cost', 'prediction_label underpredict_cost', 'over_noML10% overpredict_cost', 'over_noML20% overpredict_cost', 'over_noML30% overpredict_cost', 'under_noML10% underpredict_cost', 'under_noML20% underpredict_cost', 'under_noML30% underpredict_cost']]
display(df_bin_summary)
```
